Merging Comment Files



```python
# upload Comments.csv
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['Comments.csv']))
```

```python
# upload NEWSComments.csv
uploaded = files.upload()
```

```python
df1 = pd.read_csv(io.BytesIO(uploaded['NEWSComments.csv']))
```

```python
# concat dataframes
data = [df, df1]
comments = pd.concat(data)
comments.describe()
```

```python
# remove duplicates
commentNew = comments.drop_duplicates(subset=['commentId'])
commentNew.describe()
```

```python
# remove duplicates
comment = commentNew.drop_duplicates(subset=['text'])
comment.describe()
```

```python
comment.reset_index(inplace=True)
```

```python
comment.to_csv('AllComments.csv', index=False)
```
