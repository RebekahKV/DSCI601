```python
# Upload jsonbyline.zip from CommentsParsed/NEWSCommentsParsed to test the following code
# re-run the code for jsonbyline.zip in NEWSCommentsParsed
from zipfile import ZipFile
comments = "/content/jsonbyline.zip" # change zip file path for NEWSCommentsParsed

# Extracts the contents of the specified ZIP file.
# Parameters:
#   comments (str): The path to the ZIP file.
# Returns:
#   None
# Note:
#   This function extracts all files from the ZIP file into the current working directory.

with ZipFile(comments, 'r') as zip:
  zip.extractall()
  print('Done')
```

```python
import os
import json

def readJson(folder_path):
    """
    Reads JSON files from the specified folder path and returns a list of dictionaries.

    Args:
        folder_path (str): The path to the folder containing JSON files.

    Returns:
        list: A list of dictionaries containing data from the JSON files.
    """
    data = []
    for filename in os.listdir(folder_path):
        if filename.endswith("_jsonbyline.txt"):
            filepath = os.path.join(folder_path, filename)
            with open(filepath, "r") as file:
                jsonFile = file.readlines()
                for line in jsonFile:
                    data.append(json.loads(line))
    return data

folder_path = "/content/jsonbyline" # change folder path for NEWSCommentsParsed
data = readJson(folder_path)
```

```python
import pandas as pd
df = pd.DataFrame(data) # stores the data list in a dataframe
```

```python
# download this file to .csv
df.to_csv('Comments.csv', index=False) # change file name to 'NEWSComments.csv' for NEWSCommentsParsed
```
