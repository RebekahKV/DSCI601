# Merging Comment Files



```python
# upload Comments.csv
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['Comments.csv']))
```

```python
# upload NEWSComments.csv
uploaded = files.upload()
```

```python
df1 = pd.read_csv(io.BytesIO(uploaded['NEWSComments.csv']))
```

```python
# Concatenates multiple dataframes into a single dataframe and then describes it.
#
# Parameters:
#   data (list): A list of pandas DataFrames to concatenate.
#
# Returns:
#   pandas.DataFrame: The concatenated dataframe.
#
# Note:
#   This function concatenates the dataframes along rows (axis 0) by default.
data = [df, df1]
comments = pd.concat(data)
comments.describe()
```

```python
# Remove duplicate comments based on the 'commentId' column
# Parameters:
#   comments (DataFrame): The DataFrame containing comments.
# Returns:
#   DataFrame: A new DataFrame with duplicate comments removed based on the 'commentId' column.
# Note:
#   This function removes duplicate rows based on the specified column and retains the first occurrence of each unique commentId.
commentNew = comments.drop_duplicates(subset=['commentId'])

# Generate descriptive statistics of the new DataFrame
# Returns:
#   DataFrame: A DataFrame containing descriptive statistics of the new DataFrame.
# Note:
#   This function provides summary statistics of the new DataFrame, including count and unique.
commentNew.describe()
```

```python
# Remove duplicate comments based on the 'text' column
# Parameters:
#   comment (DataFrame): The DataFrame containing comments.
# Returns:
#   DataFrame: A new DataFrame with duplicate comments removed based on the 'text' column.
# Note:
#   This function removes duplicate rows based on the specified column and retains the first occurrence of each unique commentId.
comment = commentNew.drop_duplicates(subset=['text'])

# Generate descriptive statistics of the new DataFrame
# Returns:
#   DataFrame: A DataFrame containing descriptive statistics of the new DataFrame.
# Note:
#   This function provides summary statistics of the new DataFrame, including count and unique.
comment.describe()
```

```python
# Reset the index of the DataFrame.
#
# This method resets the index of the DataFrame. By default, it drops the current index and replaces it with a new one.
#
# Parameters:
#     inplace (bool, optional): If True, modify the DataFrame in place.
#
# Returns:
#     If inplace is True, returns None.

comment.reset_index(inplace=True)
```

```python
# stores the dataframe in a .csv file
comment.to_csv('AllComments.csv', index=False)
```
